```python
df_sorted = df.sort_values(by="2022 Population", ascending=False)
```

```python
top_10_df = df_sorted.head(10)

norm = plt.Normalize(top_10_df["2022 Population"].min(), top_10_df["2022 Population"].max())
colors = plt.cm.Reds(norm(top_10_df["2022 Population"]))

fig, ax = plt.subplots(figsize=(18, 8))
bars = ax.bar(top_10_df["Country/Territory"], top_10_df["2022 Population"], color=colors)

ax.set_xlabel("Country")
ax.set_ylabel("2022 Population (in billions)")
ax.set_title("Top 10 Most Populous Countries in 2022")

plt.tight_layout()
plt.show()
```

```python
# Now for the population distribution across the continents 
```

```python
population_by_continent = df.groupby("Continent")["2022 Population"].sum().sort_values(ascending=False)

norm = plt.Normalize(population_by_continent.min(), population_by_continent.max())
colors = plt.cm.Reds(norm(population_by_continent))

fig, ax = plt.subplots(figsize=(10, 7))
wedges, texts, autotexts = ax.pie(population_by_continent, labels=population_by_continent.index, autopct='%1.1f%%', colors=colors, startangle=140)

ax.legend(wedges, population_by_continent.index, title="Continents", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

ax.set_title("Population Distribution Across Continents in 2022")

plt.setp(autotexts, size=10, weight="bold", color="white")
plt.setp(texts, size=10)

plt.show()
```

```python
# For the Population Growth Statistics Over the Ages
```

```python
years = ["1970 Population", "1980 Population", "1990 Population", "2000 Population", "2010 Population", "2015 Population", "2020 Population", "2022 Population"]
df_continents = df.groupby("Continent")[years].sum().transpose()
```

```python
df_continents.index = [1970, 1980, 1990, 2000, 2010, 2015, 2020, 2022]
df_continents.index.name = 'Year'

plt.figure(figsize=(14, 8))
for continent in df_continents.columns:
    plt.plot(df_continents.index, df_continents[continent], marker='o', label=continent)

plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Population Timeline of All Continents")
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
# Countries with Highest Growth Rate
```

```python
df_sorted = df.sort_values(by="Growth Rate", ascending=False)
```

```python
top_10_df = df_sorted.head(10)

norm = plt.Normalize(top_10_df["Growth Rate"].min(), top_10_df["Growth Rate"].max())
colors = plt.cm.Greens(norm(top_10_df["Growth Rate"]))

fig, ax = plt.subplots(figsize=(14, 8))
bars = ax.bar(top_10_df["Country/Territory"], top_10_df["Growth Rate"], color=colors)

ax.set_xlabel("Country")
ax.set_ylabel("Growth Rate")
ax.set_title("Top 10 Countries with the Highest Population Growth Rate")

plt.xticks(rotation=25, ha='right')

plt.tight_layout()

plt.show()
```

```python
# Countries with least Population Growth Rate
```

```python
df_sorted = df.sort_values(by="Growth Rate", ascending=True)
```

```python
last_10_df = df_sorted.head(10)

norm = plt.Normalize(last_10_df["Growth Rate"].min(), last_10_df["Growth Rate"].max())
colors = plt.cm.Greens(norm(last_10_df["Growth Rate"]))

fig, ax = plt.subplots(figsize=(14, 8))
bars = ax.bar(last_10_df["Country/Territory"], last_10_df["Growth Rate"], color=colors)

ax.set_xlabel("Country")
ax.set_ylabel("Growth Rate")
ax.set_title("Top 10 Countries with the Lowest Population Growth Rate")

plt.xticks(rotation=25, ha='right')

plt.tight_layout()

plt.show()
```

```python
# Continents with highest average growth rate
```

```python
avg_growth_rate = df.groupby("Continent")["Growth Rate"].mean().sort_values(ascending=False)

# Plot the average growth rate for each continent
fig, ax = plt.subplots(figsize=(12, 8))
bars = ax.bar(avg_growth_rate.index, avg_growth_rate, color=plt.cm.Greens(np.linspace(0.2, 0.8, len(avg_growth_rate))))

# Add labels and title
ax.set_xlabel("Continent")
ax.set_ylabel("Average Growth Rate")
ax.set_title("Average Population Growth Rate by Continent")
ax.bar_label(bars, fmt='%.4f')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=25, ha='right')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
